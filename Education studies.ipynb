```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
from datasets import load_dataset
from huggingface_hub import hf_hub_download
import json

# Load the dataset
dataset = load_dataset("socratesft/SocSci210") 

# Load a mapping file
mapping_file = hf_hub_download(
    repo_id="socratesft/SocSci210",
    filename="metadata/participant_mapping.json",
    repo_type="dataset"
)

with open(mapping_file, 'r') as f:
    participant_mapping = json.load(f)
```

```python
def convertToPandas(dataset,setName):
    return dataset[setName].to_pandas()
```

```python
datasetTrainPS = convertToPandas(dataset,"train")
datasetTrainPS.head()
```

```python
studyIdValues = set()
for studyValue in datasetTrainPS["study_id"]:
    studyIdValues.add(studyValue)
    
print(len(studyIdValues))
```

```python
datasetTrainPS["demographic"]
```

```python
education_dataframe = pd.concat([
    datasetTrainPS[["study_id"]].reset_index(drop=True),
    pd.DataFrame(datasetTrainPS["demographic"].tolist())
], axis=1)
```

```python
education_dataframe.head()
```

```python
#checking for nan values
print(education_dataframe["education"].isna().sum())
#removing them
education_dataframe = education_dataframe.dropna(
    subset=["education"]
)
```

```python
values = set()
#education_dataframe["education"] = education_dataframe.dropna(subset=["education"], inplace=True)
valueRow = 0
totalRows = 0
for educationValue in education_dataframe["education"]:
    values.add(educationValue)
    if educationValue == "8.0" or educationValue == "14.0":
        print("found at row: ", valueRow)
        totalRows += 1
    valueRow += 1
    
print(totalRows)
```

```python
print(values)
```

```python
education_dataframe = education_dataframe[education_dataframe['education'] != "8.0"]
education_dataframe = education_dataframe[education_dataframe['education'] != "14.0"]
```

```python
education_dataframe.head()
```

```python
studyIdValues = set()
for studyValue in education_dataframe["study_id"]:
    studyIdValues.add(studyValue)
```

```python
len(studyIdValues)
```

```python
educationalMapValue = {"Bachelor's degree":4, 
                       "Less than high school": 0, 
                       "High school graduate or equivalent":2, 
                       "Some high school (no diploma)": 1, 
                       "Post grad study/professional degree":5, 
                       "Vocational/tech school/some college/associates":3
}

education_dataframe["education_num_representation"] = (
    education_dataframe["education"].map(educationalMapValue)
)
    
```

```python
education_dataframe.head()
```

```python
cols = 4
rows = math.ceil(len(studyIdValues) / cols)

fig, axes = plt.subplots(
    rows,
    cols,
    figsize=(20, rows * 10)
)

axes = axes.flatten()

for i, study_id in enumerate(studyIdValues):
    study_dataframe = education_dataframe[education_dataframe["study_id"] == study_id]
    education_counts = (
        study_dataframe["education_num_representation"]
        .value_counts()
        .reindex(range(6), fill_value=0)
    )
    education_counts.index = [
        "<HS",
        "SHS",
        "HS",
        "College",
        "BA",
        "PG"
    ]
    axes[i].set_title(study_id)
    axes[i].set_xlabel("education")
    axes[i].set_ylabel("amount")
    education_counts.plot.bar(
        ax=axes[i]
    )
    
    plt.xlabel("education")
    plt.ylabel("amount")

# remove unused plots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.show()
```
